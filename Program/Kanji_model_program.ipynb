# แบบจำลองทางคณิตศาสตร์การเรียนรู้ตัวอักษรคันจิ
นายณัฐวุฒิ เทศงามถ้วน  116410901009-3 (ผู้หาข้อมูล)

นายคณัสนันท์ ทรัพย์อุดม  116410901033-3 (ผู้ทำโปรแกรม)

นายธนาคาร ธิคุณ  116410901048-1 (ผู้นำเสนอ)

```python
import matplotlib.pyplot as plt
import math 
import ipywidgets as widgets
from IPython.display import display, clear_output
#Normal Setting
l = 786
#Display Section
alpha = widgets.FloatText(description='𝛼')
beta = widgets.FloatText(description='𝛽')
time = widgets.IntText(description='t')
kanji_01 = widgets.IntText(description='k_01')
kanji_02 = widgets.IntText(description='k_02')
kanji_03 = widgets.IntText(description='k_03')
kanji_04 = widgets.IntText(description='k_04')
kanji_05 = widgets.IntText(description='k_05')
kanji_06 = widgets.IntText(description='k_06')

#Function Section
def L_1(t,𝛼,𝛽,K_01):
    I_1 = 80 + l
    if t <= 120:
        learning_j1 = math.floor((1 - 𝛽)*(K_01 + 𝛼*t))
        return min(I_1, learning_j1)
    elif t > 120 :
        learning_j1 = math.floor((1 - 𝛽)*(K_01 + 𝛼*120))
        return min(I_1, learning_j1)

def L_2(t,𝛼,𝛽,K_02):
    I_2 = 120 + (l - L_1(time.value,alpha.value,beta.value,kanji_01.value) + 80)
    if t <= 120:
        return 0
    elif (t > 120) and (t <= 240):
        learning_j2 = math.floor((1 - 𝛽)*(K_02 + 𝛼*(t-120)))
        return min(I_2, learning_j2)
    elif t > 240 :
        learning_j2 = math.floor((1 - 𝛽)*(K_02 + 𝛼*120))
        return min(I_2, learning_j2)

def L_3(t,𝛼,𝛽,K_03):
    I_3 = 250 + (l - L_1(time.value,alpha.value,beta.value,kanji_01.value) + 80 
             - L_2(time.value,alpha.value,beta.value,kanji_02.value) + 120)
    if t <= 240:
        return 0
    elif (t > 240) and (t <= 360):
        learning_j3 = math.floor((1 - 𝛽)*(K_03 + 𝛼*(t-240)))
        return min(I_3, learning_j3)
    elif t > 360 :
        learning_j3 = math.floor((1 - 𝛽)*(K_03 + 𝛼*120))
        return min(I_3, learning_j3)

def L_4(t,𝛼,𝛽,K_04):
    I_4 = 250 + (l - L_1(time.value,alpha.value,beta.value,kanji_01.value) + 80 
             - L_2(time.value,alpha.value,beta.value,kanji_02.value) + 120 
             - L_3(time.value,alpha.value,beta.value,kanji_03.value) + 250)
    if t <= 360:
        return 0
    elif (t > 360) and (t <= 480):
        learning_j4 = math.floor((1 - 𝛽)*(K_04 + 𝛼*(t-360)))
        return min(I_4, learning_j4)
    elif t > 480 :
        learning_j4 = math.floor((1 - 𝛽)*(K_04 + 𝛼*120))
        return min(I_4, learning_j4)

def L_5(t,𝛼,𝛽,K_05):
    I_5 = 300 + (l - L_1(time.value,alpha.value,beta.value,kanji_01.value) + 80 
             - L_2(time.value,alpha.value,beta.value,kanji_02.value) + 120 
             - L_3(time.value,alpha.value,beta.value,kanji_03.value) + 250 
             - L_4(time.value,alpha.value,beta.value,kanji_04.value) + 250)
    if t <= 480:
        return 0
    elif (t > 480) and (t <= 600):
        learning_j5 = math.floor((1 - 𝛽)*(K_05 + 𝛼*(t-480)))
        return min(I_5, learning_j5)
    elif t > 600 :
        learning_j5 = math.floor((1 - 𝛽)*(K_05 + 𝛼*120))
        return min(I_5, learning_j5)

def L_6(t,𝛼,𝛽,K_06):
    I_6 = 350 + (l - L_1(time.value,alpha.value,beta.value,kanji_01.value) + 80 
             - L_2(time.value,alpha.value,beta.value,kanji_02.value) + 120 
             - L_3(time.value,alpha.value,beta.value,kanji_03.value) + 250 
             - L_4(time.value,alpha.value,beta.value,kanji_04.value) + 250 
             - L_5(time.value,alpha.value,beta.value,kanji_05.value) + 300)
    if t <= 600:
        return 0
    elif (t > 600) and (t <= 720):
        learning_j6 = math.floor((1 - 𝛽)*(K_06 + 𝛼*(t-600)))
        return min(I_6, learning_j6)
    elif t > 720 :
        learning_j6 = math.floor((1 - 𝛽)*(K_06 + 𝛼*120))
        return min(I_6, learning_j6)
#Solution Check
def fine_alpha(𝛽,K_01,K_02,K_03,K_04,K_05,K_06):
    𝛼_star = (2136 - (K_01 + K_02 + K_03 + K_04 + K_05 + K_06)*(1 - 𝛽))/(720*(1 - 𝛽))
    return 𝛼_star
def out_range_fine_alpha(𝛽,K_01,K_02,K_03,K_04,K_05,K_06):
    𝛼_star = (2137 - (K_01 + K_02 + K_03 + K_04 + K_05 + K_06)*(1 - 𝛽))/(720*(1 - 𝛽))
    return 𝛼_star
#Button click event handler
def calculate(btn):
    with output:
        clear_output() 
        display(f'Japanise 1 Kanji Learned : {L_1(time.value,alpha.value,beta.value,kanji_01.value)} Character')
        display(f'Japanise 2 Kanji Learned : {L_2(time.value,alpha.value,beta.value,kanji_02.value)} Character')
        display(f'Japanise 3 Kanji Learned : {L_3(time.value,alpha.value,beta.value,kanji_03.value)} Character')
        display(f'Japanise 4 Kanji Learned : {L_4(time.value,alpha.value,beta.value,kanji_04.value)} Character')
        display(f'Japanise 5 Kanji Learned : {L_5(time.value,alpha.value,beta.value,kanji_05.value)} Character')
        display(f'Japanise 6 Kanji Learned : {L_6(time.value,alpha.value,beta.value,kanji_06.value)} Character')
        display(plot(time.value, alpha.value, beta.value, 
                     kanji_01.value, kanji_02.value, kanji_03.value, kanji_04.value, kanji_05.value, kanji_06.value))
        display(f'The fine alpha is : {fine_alpha(beta.value,kanji_01.value,kanji_02.value,kanji_03.value,kanji_04.value,kanji_05.value,kanji_06.value)} ≤ 𝛼 < {out_range_fine_alpha(beta.value,kanji_01.value,kanji_02.value,kanji_03.value,kanji_04.value,kanji_05.value,kanji_06.value)}')
#Plot Section
def plot(t,𝛼,𝛽,K_01,K_02,K_03,K_04,K_05,K_06) :
    list_Lt = []
    list_t = []
    for i in range(0,t+1):
        list_t.append(i)
    
        Lt = L_1(i,𝛼,𝛽,K_01) + L_2(i,𝛼,𝛽,K_02) + L_3(i,𝛼,𝛽,K_03) + L_4(i,𝛼,𝛽,K_04) + L_5(i,𝛼,𝛽,K_05) + L_6(i,𝛼,𝛽,K_06)
        list_Lt.append(Lt)
    plt.plot(list_t, list_Lt, label = 'L(t)')
    plt.xlabel('Days')
    plt.ylabel('Kanji learned')
    plt.legend()
    plt.grid()
    plt.show()
    return f'All Kanji Learned is {list_Lt[-1]} Character'

#User Interface Section
output = widgets.Output()
calc_button = widgets.Button(description='Calculate')
calc_button.on_click(calculate)

ui1 = widgets.VBox([alpha, beta, time, kanji_01, kanji_02, kanji_03, kanji_04, kanji_05, kanji_06])
outer = widgets.VBox([calc_button,output])
full_screen = widgets.HBox([ui1,outer])
display(full_screen)
```
