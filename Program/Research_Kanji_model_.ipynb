```python
import math 
#‡∏û‡∏¥‡∏à‡∏≤‡∏£‡∏ì‡∏≤‡πÇ‡∏à‡∏ó‡∏¢‡πå‡∏à‡∏∞‡πÑ‡∏î‡πâ ‡∏™‡∏°‡∏Å‡∏≤‡∏£ ‡∏™‡∏≥‡∏´‡∏£‡∏±‡∏ö‡πÇ‡∏°‡πÄ‡∏î‡∏•‡∏≠‡∏¢‡πà‡∏≤‡∏á‡∏á‡πà‡∏≤‡∏¢‡∏î‡∏±‡∏á‡∏ô‡∏µ‡πâ
#‡∏™‡πà‡∏ß‡∏ô‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£
K_0 = 100
r = 10 #learning rate
t = 10 #time
limit = 160 
#‡∏™‡πà‡∏ß‡∏ô‡πÇ‡∏°‡πÄ‡∏î‡∏•‡πÅ‡∏ö‡∏ö‡∏á‡πà‡∏≤‡∏¢
K_t = K_0 + r*t
print(K_t)

#‡∏™‡πà‡∏ß‡∏ô‡∏•‡∏≠‡∏á‡∏´‡∏≤‡∏Ñ‡∏≥‡∏ï‡∏≠‡∏ö‡∏ß‡πà‡∏≤‡∏à‡∏∞‡πÉ‡∏ä‡πâ‡πÄ‡∏ß‡∏•‡∏≤‡∏Å‡∏µ‡πà‡∏ß‡∏±‡∏ô‡∏ñ‡∏∂‡∏á‡∏à‡∏∞‡πÄ‡∏£‡∏µ‡∏¢‡∏ô‡∏£‡∏π‡πâ‡∏Ñ‡∏±‡∏ô‡∏à‡∏¥‡πÑ‡∏î‡πâ‡∏Ñ‡∏£‡∏ö 2136 
for times in range(1, limit+1):
    K_t = min(K_0 + r*times, 2136)
    if K_t == 2136:
        print(f'‡∏ï‡πâ‡∏≠‡∏á‡πÉ‡∏ä‡πâ‡πÄ‡∏ß‡∏•‡∏≤‡πÄ‡∏£‡∏µ‡∏¢‡∏ô {t} ‡∏ß‡∏±‡∏ô ‡∏ñ‡∏∂‡∏á‡∏à‡∏∞‡∏à‡∏î‡∏à‡∏≥‡∏Ñ‡∏±‡∏ô‡∏à‡∏¥‡πÑ‡∏î‡πâ‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î 2136 ‡∏ï‡∏±‡∏ß')
        print(f'‡∏Ñ‡∏¥‡∏î‡πÄ‡∏õ‡πá‡∏ô‡πÄ‡∏î‡∏∑‡∏≠‡∏ô‡∏Å‡πá‡∏Ñ‡∏∑‡∏≠‡∏ï‡πâ‡∏≠‡∏á‡πÉ‡∏ä‡πâ‡πÄ‡∏ß‡∏•‡∏≤ {t//30} ‡πÄ‡∏î‡∏∑‡∏≠‡∏ô {math.floor(t - (t//30)*30)} ‡∏ß‡∏±‡∏ô ‡∏ñ‡∏∂‡∏á‡∏à‡∏∞‡∏à‡∏î‡∏à‡∏≥‡∏Ñ‡∏±‡∏ô‡∏à‡∏¥‡πÑ‡∏î‡πâ‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î 2136 ‡∏ï‡∏±‡∏ß')
        break
    elif times == limit and K_t < 2136:
        print(f'‡πÅ‡∏°‡πâ‡∏ß‡πà‡∏≤‡∏à‡∏∞‡πÉ‡∏ä‡πâ‡πÄ‡∏ß‡∏•‡∏≤‡∏Ñ‡∏£‡∏ö {limit} ‡∏ß‡∏±‡∏ô‡πÅ‡∏•‡πâ‡∏ß ‡∏ú‡∏π‡πâ‡∏®‡∏∂‡∏Å‡∏©‡∏≤‡∏Ñ‡∏±‡∏ô‡∏à‡∏¥‡∏¢‡∏±‡∏á‡∏Ñ‡∏á‡∏à‡∏î‡∏à‡∏≥‡∏Ñ‡∏≥‡∏à‡∏¥‡πÑ‡∏î‡πâ‡πÄ‡∏û‡∏µ‡∏¢‡∏á {K_t} ‡∏ï‡∏±‡∏ß')
        break
```

# <span style="color:blue">The Mathematical Model Part : Linear Equation </span> 
‡∏à‡∏≤‡∏Å‡∏Å‡∏≤‡∏£‡∏û‡∏¥‡∏à‡∏≤‡∏£‡∏ì‡∏≤‡πÇ‡∏à‡∏ó‡∏¢‡πå ‡∏ó‡∏≥‡πÉ‡∏´‡πâ‡πÄ‡∏£‡∏≤‡∏û‡∏ö‡∏™‡∏¥‡πà‡∏á‡∏ó‡∏µ‡πà‡∏û‡∏ß‡∏Å‡πÄ‡∏Ç‡∏≤‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡∏Ñ‡∏∑‡∏≠ ‡∏ï‡πâ‡∏≠‡∏á‡∏Å‡∏≤‡∏£‡∏ó‡∏£‡∏≤‡∏ö‡∏ß‡πà‡∏≤‡∏à‡∏∞‡πÉ‡∏ä‡πâ‡πÄ‡∏ß‡∏•‡∏≤‡∏Å‡∏µ‡πà‡∏ß‡∏±‡∏ô‡∏ñ‡∏∂‡∏á‡∏à‡∏∞‡πÄ‡∏£‡∏µ‡∏¢‡∏ô‡∏£‡∏π‡πâ‡∏Ñ‡∏±‡∏ô‡∏à‡∏¥‡∏ó‡∏±‡πâ‡∏á 2,136 ‡∏ï‡∏±‡∏ß‡πÑ‡∏î‡πâ ‡πÇ‡∏î‡∏¢‡πÄ‡∏£‡∏≤‡∏ï‡∏±‡πâ‡∏á‡∏õ‡∏£‡∏∞‡πÄ‡∏î‡πá‡∏ô‡∏´‡∏•‡∏±‡∏Å‡πÑ‡∏õ‡∏ó‡∏µ‡πà‡∏ï‡∏±‡∏ß‡∏Ñ‡∏±‡∏ô‡∏à‡∏¥‡∏î‡∏±‡πà‡∏á‡∏™‡∏°‡∏Å‡∏≤‡∏£‡∏ï‡πà‡∏≠‡πÑ‡∏õ‡∏ô‡∏µ‡πâ <br>
‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡πÉ‡∏´‡πâ $ K_0 , t \in \mathbb{N}_+ $ ‡πÅ‡∏•‡∏∞ $ \alpha , \beta \in \mathbb{R}_+ $
$$
\begin{aligned}
Kl(t) & = K_0 + ùõºt \\
r(t) & =  \beta t  \\
K(t) & = \left\lfloor Kl(t) - r(t) \right\rfloor \\
K(t) & = \left\lfloor K_0 + ùõºt - \beta t \right\rfloor
\end{aligned}
$$
‡πÇ‡∏î‡∏¢‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏ï‡∏±‡∏ß‡∏°‡∏µ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏´‡∏°‡∏≤‡∏¢‡∏î‡∏±‡∏á‡∏ô‡∏µ‡πâ <br>
Kl(t) =  Kanji Learned : ‡∏Ñ‡∏±‡∏ô‡∏à‡∏¥‡∏ó‡∏µ‡πà‡∏ú‡∏π‡πâ‡∏®‡∏∂‡∏Å‡∏©‡∏≤‡∏Ñ‡∏±‡∏ô‡∏à‡∏¥‡πÄ‡∏£‡∏µ‡∏¢‡∏ô‡πÅ‡∏•‡∏∞‡∏à‡∏≥‡πÑ‡∏î‡πâ‡∏≠‡∏¢‡∏π‡πà (‡∏ï‡∏±‡∏ß ‡∏´‡∏£‡∏∑‡∏≠ Â≠ó)<br>
r(t) = Reduce by losing some kanji : ‡∏Ñ‡∏±‡∏ô‡∏à‡∏¥‡∏ó‡∏µ‡πà‡∏ú‡∏π‡πâ‡∏®‡∏∂‡∏Å‡∏©‡∏≤‡∏à‡∏∞‡∏™‡∏π‡∏ç‡πÄ‡∏™‡∏µ‡∏¢‡πÑ‡∏õ‡∏´‡∏•‡∏±‡∏á‡∏®‡∏∂‡∏Å‡∏©‡∏≤‡∏Ñ‡∏±‡∏ô‡∏à‡∏¥‡πÉ‡∏ô‡∏õ‡∏£‡∏¥‡∏°‡∏≤‡∏ì‡∏°‡∏≤‡∏Å‡∏Ç‡∏≠‡∏á‡∏ß‡∏±‡∏ô‡πÜ ‡∏ô‡∏±‡πâ‡∏ô (‡∏ï‡∏±‡∏ß ‡∏´‡∏£‡∏∑‡∏≠ Â≠ó)<br>
K(t) = Kanji : ‡∏Ñ‡∏±‡∏ô‡∏à‡∏¥‡∏ó‡∏µ‡πà‡∏ú‡∏π‡πâ‡∏®‡∏∂‡∏Å‡∏©‡∏≤‡∏Ñ‡∏±‡∏ô‡∏à‡∏¥‡∏à‡∏≥‡πÑ‡∏î‡πâ‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î
K_0 = Starter Kanji : ‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£‡∏Ñ‡∏±‡∏ô‡∏à‡∏¥‡∏ó‡∏µ‡πà‡∏ú‡∏π‡πâ‡∏®‡∏∂‡∏Å‡∏©‡∏≤‡∏ï‡∏±‡∏ß‡∏Ñ‡∏±‡∏ô‡∏à‡∏¥‡∏£‡∏π‡πâ (‡∏ï‡∏±‡∏ß ‡∏´‡∏£‡∏∑‡∏≠ Â≠ó)<br>
(‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡πÉ‡∏´‡πâ K_0  ‡∏≠‡∏¢‡∏π‡πà‡∏ó‡∏µ‡πà 214 ‡∏ï‡∏±‡∏ß) <br>
t = Time : ‡πÄ‡∏ß‡∏•‡∏≤ ‡∏´‡∏£‡∏∑‡∏≠ ‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡∏®‡∏∂‡∏Å‡∏©‡∏≤‡∏Ñ‡∏±‡∏ô‡∏à‡∏¥ (‡∏ß‡∏±‡∏ô) <br>
$ \alpha $ = $A$ learning kanji rate : ‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£‡∏Ñ‡∏±‡∏ô‡∏à‡∏¥‡∏ó‡∏µ‡πà‡∏ú‡∏π‡πâ‡∏®‡∏∂‡∏Å‡∏©‡∏≤‡∏Ñ‡∏±‡∏ô‡∏à‡∏¥ ‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ‡∏®‡∏∂‡∏Å‡∏©‡∏≤‡πÑ‡∏î‡πâ‡∏ï‡πà‡∏≠‡∏ß‡∏±‡∏ô (‡∏ï‡∏±‡∏ß/‡∏ß‡∏±‡∏ô ‡∏´‡∏£‡∏∑‡∏≠ Â≠ó/‡∏ß‡∏±‡∏ô) <br>
$ \beta $ = $B$eing reduce by Forgetting Curve : ‡∏≠‡∏±‡∏ï‡∏£‡∏≤‡∏Å‡∏≤‡∏£‡∏™‡∏π‡∏ç‡πÄ‡∏™‡∏µ‡∏¢‡∏ï‡∏±‡∏ß‡∏Ñ‡∏±‡∏ô‡∏à‡∏¥‡∏Ç‡∏≠‡∏á‡∏ú‡∏π‡πâ‡∏®‡∏∂‡∏Å‡∏©‡∏≤‡∏Ñ‡∏±‡∏ô‡∏à‡∏¥ (‡∏ï‡∏±‡∏ß/‡∏ß‡∏±‡∏ô ‡∏´‡∏£‡∏∑‡∏≠ Â≠ó/‡∏ß‡∏±‡∏ô)<br>
(‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡πÉ‡∏´‡πâ $\beta$ ‡∏à‡∏≤‡∏Å Forgetting Curve ‡∏≠‡∏¢‡∏π‡πà‡∏ó‡∏µ‡πà 0.5 ‡∏´‡∏£‡∏∑‡∏≠ $\frac{5}{10}$)

```python
import math 
import ipywidgets as widgets
import matplotlib.pyplot as plt
import numpy as np
#‡∏™‡πà‡∏ß‡∏ô‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£
K_0 = 100
ùõº = 10
ùõΩ = ùõº*1/2
#‡∏™‡πà‡∏ß‡∏ô‡∏ü‡∏±‡∏á‡∏Å‡πå‡∏ä‡∏±‡∏ô
def Al(t):
    all_kanji = K_0 + ùõº*t
    return(all_kanji)
def r(t):
    reduce_rate = ùõΩ*t
    return(reduce_rate)
def K(t):
    now_kanji = math.ceil(Al(t) - r(t))
    return min(now_kanji , 2136)
def LK(t):
    learner_kanji = math.ceil(2136 + ùõΩ*t - K_0 - ùõº*t)
    return max(learner_kanji , 0)
#‡∏™‡πà‡∏ß‡∏ô plot
list_t = []
list_LK = []
list_K = []
for t in range(1,10+1):
    list_t.append(t)
    
    LKt = LK(t)
    list_LK.append(LKt)
    
    Kt = K(t)
    list_K.append(Kt)
plt.plot(list_t, list_LK, label = 'LK(t)')
plt.plot(list_t, list_K, label = 'K(t)')
plt.xlabel('Days')
plt.ylabel('Kanji Lefts')
plt.legend()
plt.grid()
plt.show()
print(f'‡∏ï‡∏≠‡∏ô‡∏ô‡∏µ‡πâ‡∏ú‡∏π‡πâ‡∏®‡∏∂‡∏Å‡∏©‡∏≤‡∏Ñ‡∏±‡∏ô‡∏à‡∏¥ ‡πÄ‡∏´‡∏•‡∏∑‡∏≠‡∏Ñ‡∏±‡∏ô‡∏à‡∏¥‡∏ó‡∏µ‡πà‡∏ï‡πâ‡∏≠‡∏á‡∏®‡∏∂‡∏Å‡∏©‡∏≤‡∏≠‡∏¢‡∏π‡πà‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î {list_K[-1]} ‡∏ï‡∏±‡∏ß')
print(f'‡∏î‡∏±‡∏á‡∏ô‡∏±‡πâ‡∏ô‡∏ï‡∏≠‡∏ô‡∏ô‡∏µ‡πâ‡∏ú‡∏π‡πâ‡∏®‡∏∂‡∏Å‡∏©‡∏≤‡∏Ñ‡∏±‡∏ô‡∏à‡∏¥ ‡πÄ‡∏£‡∏µ‡∏¢‡∏ô‡∏£‡∏π‡πâ‡∏Ñ‡∏±‡∏ô‡∏à‡∏¥‡πÑ‡∏î‡πâ‡∏ó‡∏±‡πâ‡∏á‡∏´‡∏°‡∏î {list_LK[-1]} ‡∏ï‡∏±‡∏ß')
print(list_LK)
```

# <span style="color:blue">The Mathematical Model Part : Quadratic Equation </span> 
‡∏Ñ‡∏£‡∏≤‡∏ß‡∏ô‡∏µ‡πâ‡πÄ‡∏£‡∏≤‡∏°‡∏≤‡∏ó‡∏≥‡∏ï‡∏±‡∏ß‡∏™‡∏°‡∏Å‡∏≤‡∏£‡∏ó‡∏µ‡πà‡∏à‡∏∞‡πÄ‡∏≠‡∏≤‡πÑ‡∏õ‡∏ô‡∏≥‡πÄ‡∏™‡∏ô‡∏≠‡∏Å‡∏±‡∏ô ‡πÇ‡∏î‡∏¢‡∏à‡∏∞‡∏°‡∏µ‡∏´‡∏ô‡πâ‡∏≤‡∏ï‡∏≤‡∏™‡∏°‡∏Å‡∏≤‡∏£‡∏î‡∏±‡∏á‡∏ô‡∏µ‡πâ <br>
‡∏Å‡∏≥‡∏´‡∏ô‡∏î‡πÉ‡∏´‡πâ $ K_0 , t \in \mathbb{N}_+ $ ‡πÅ‡∏•‡∏∞ $ \alpha , \beta \in \mathbb{R}_+ $
$$ K(t) = 2136 - \left\lfloor K_0 + \alpha t - \beta \alpha t^2 \right\rfloor $$
$$ L(t) = \left\lfloor K_0 + \alpha t - \beta \alpha t^2 \right\rfloor $$
$\textbf{‡πÇ‡∏î‡∏¢‡∏ï‡∏±‡∏ß‡πÅ‡∏õ‡∏£‡πÅ‡∏ï‡πà‡∏•‡∏∞‡∏ï‡∏±‡∏ß ‡πÅ‡∏ó‡∏ô‡∏Ñ‡∏ß‡∏≤‡∏°‡∏´‡∏°‡∏≤‡∏¢‡∏î‡∏±‡∏á‡∏ô‡∏µ‡πâ}$ <br>
K(t) ‡πÅ‡∏ó‡∏ô ‡πÇ‡∏à‡πÇ‡∏¢‡∏Ñ‡∏±‡∏ô‡∏à‡∏¥‡∏ó‡∏µ‡πà‡πÄ‡∏´‡∏•‡∏∑‡∏≠‡∏ó‡∏µ‡πà‡∏ï‡πâ‡∏≠‡∏á‡πÄ‡∏£‡∏µ‡∏¢‡∏ô (‡∏ï‡∏±‡∏ß) <br>
L(t) ‡πÅ‡∏ó‡∏ô ‡∏Ñ‡∏±‡∏ô‡∏à‡∏¥‡∏ó‡∏µ‡πà‡∏ú‡∏π‡πâ‡∏®‡∏∂‡∏Å‡∏©‡∏≤‡∏£‡∏π‡πâ ‡∏ì ‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà t (‡∏ï‡∏±‡∏ß) <br>
K_0 ‡πÅ‡∏ó‡∏ô ‡∏£‡∏≤‡∏Å‡∏®‡∏±‡∏û‡∏ó‡πå‡∏ó‡∏µ‡πà‡∏à‡∏≥‡πÄ‡∏õ‡πá‡∏ô‡∏ï‡πà‡∏≠‡∏Å‡∏≤‡∏£‡πÄ‡∏Ç‡πâ‡∏≤‡πÉ‡∏à‡∏ï‡∏±‡∏ß‡∏Ñ‡∏±‡∏ô‡∏à‡∏¥ 214 ‡∏ï‡∏±‡∏ß‡∏≠‡∏±‡∏Å‡∏©‡∏£ (‡∏ï‡∏±‡∏ß) <br>
$ \alpha $ ‡πÅ‡∏ó‡∏ô ‡∏Ñ‡∏ß‡∏≤‡∏°‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ‡πÉ‡∏ô‡∏Å‡∏≤‡∏£‡πÄ‡∏£‡∏µ‡∏¢‡∏ô‡∏£‡∏π‡πâ‡∏Ñ‡∏±‡∏ô‡∏à‡∏¥ (‡∏ï‡∏±‡∏ß/‡∏ß‡∏±‡∏ô) <br>
t ‡πÅ‡∏ó‡∏ô ‡∏ß‡∏±‡∏ô‡∏ó‡∏µ‡πà‡πÄ‡∏£‡∏µ‡∏¢‡∏ô‡∏Ñ‡∏±‡∏ô‡∏à‡∏¥ (‡∏ß‡∏±‡∏ô) <br>
$ \beta $ ‡πÅ‡∏ó‡∏ô ‡∏≠‡∏±‡∏ï‡∏£‡∏≤‡∏Å‡∏≤‡∏£‡∏™‡∏π‡∏ç‡πÄ‡∏™‡∏µ‡∏¢‡∏Ñ‡∏±‡∏ô‡∏à‡∏¥‡∏ï‡∏≤‡∏°‡∏´‡∏•‡∏±‡∏Å‡∏Å‡∏≤‡∏£ Forgetting Curve [‡∏ï‡∏≤‡∏°‡∏™‡∏ñ‡∏¥‡∏ï‡∏¥ ‡∏Ñ‡∏∑‡∏≠ 66.3% ‡∏ï‡πà‡∏≠‡∏ß‡∏±‡∏ô] (‡πÑ‡∏°‡πà‡∏°‡∏µ‡∏´‡∏ô‡πà‡∏ß‡∏¢) <br>
$ \gamma $ ‡πÅ‡∏ó‡∏ô ‡∏≠‡∏±‡∏ï‡∏£‡∏≤‡πÄ‡∏£‡πà‡∏á‡∏Å‡∏≤‡∏£‡∏™‡∏π‡∏ç‡πÄ‡∏™‡∏µ‡∏¢‡∏Ñ‡∏±‡∏ô‡∏à‡∏¥‡πÅ‡∏ö‡∏ö‡∏™‡∏∞‡∏™‡∏° (‡∏ï‡∏±‡∏ß)

‡πÄ‡∏°‡∏∑‡πà‡∏≠ $ \alpha $ ‡∏™‡∏≤‡∏°‡∏≤‡∏£‡∏ñ‡∏´‡∏≤‡πÑ‡∏î‡πâ‡∏à‡∏≤‡∏Å <br> 
$$
\begin{aligned}
2136 & = \left\lfloor 214 + 360\alpha  - 360(0.663^2)\alpha  \right\rfloor \\
2136 & \leq 214 + 360\alpha  - 360(0.663^2)\alpha < 2137 \\
1922 & \leq 360\alpha  - 360(0.663^2)\alpha < 1923 \\
\frac{1922}{360} & \leq \alpha - 0.663^2\alpha < \frac{1923}{360} \\
\frac{1922}{360} & \leq \alpha - 0.439569\alpha < \frac{1923}{360}
